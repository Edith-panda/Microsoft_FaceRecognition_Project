```python
from google.colab.patches import cv2
# importing the haarcascade frontalface which contains algorithm used for faces in image or real time video.
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascade_frontalface_default.xml')
img = cv2.imread('/content/drive/MyDrive/test5.jpg')


# converting the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


#setting the dimensions for face detection
faces = face_cascade.detectMultiScale(gray, 1.1 ,4)
for (x, y , w , h) in faces:
    cv2.rectangle(img, (x , y ),(x+w , y+h),(0, 0, 255) , 2)


# for displaying output
cv2_imshow(img)

#waiting until the next command 
cv2.waitKey()
```
